```python
from Game import Game
from Maps import *
from HER.DQNAgentWithHER import DQN_HER_params
import random
random.seed(0)

"""
  This function kick starts a particular experiment/game using DQN with
  Hindsight Experience Replay.
  In this case we choose a learning rate = 0.9 and gamma = 0.6.
  The epsilon is decaying.
"""
if __name__ == '__main__':
    params = DQN_HER_params()
    params.learning_rate = 0.9
    params.gamma = 0.6
    params.epsilon = 0.9
    params.batch_size = 128
    params.decay_epsilon = 0.999

    game = Game(policy="her",
                params=params,
                num_episodes=1000,
                num_steps_per_episode=600,
                mapp=small_map,
                num_ghosts=1,
                verbose=False)
    game.run()
    game.report_results()
```
